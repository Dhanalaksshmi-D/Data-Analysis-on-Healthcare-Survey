```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('music.csv')
data.head(21)
```

```python
#AGE DISTRIBUTION
data['1.Age group'].value_counts().plot(kind = "bar", color= 'ORANGE' )
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.show()
```

```python
#GENDER DISTFIBUTION
plt.figure(figsize=(6, 4))
data['2.Gender'].value_counts().plot(
    kind='pie',
    color='skyblue',
    autopct='%1.1f%%'
)
plt.title('gender  Distribution')
plt.show()
```

```python
# OCCUPATION DISTRIBUTION
plt.figure(figsize=(6, 4))
data['3.Occupation'].value_counts().plot(
    kind='bar',
    color='yellow',
    edgecolor='black'

)
plt.title('occupation   Distribution')
plt.xlabel('Occupation')
plt.ylabel('Number of Respondents')
plt.show()
```

```python

# --- Bar Chart: How often do people see healthcare ads ---
data['4.How often do you come across healthcare advertisements?'].value_counts().plot(kind='bar', color='green', edgecolor='black')
plt.title('Frequency of Encountering Healthcare Advertisements')
plt.xlabel('Frequency')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()
```

```python

#  WHERE PEOPLE MOSTLY SEE OR HEAR ADS

data['5.Where do you mostly see or hear healthcare marketing?'].value_counts().plot(
    kind='bar',
    color='cornflowerblue',
    edgecolor='black'
)
plt.title('Where People Mostly See or Hear Healthcare Marketing')
plt.xlabel('Platform')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

A crosstab (short for cross-tabulation) is a two-way table that shows the relationship between two categorical variables — like a mini summary or comparison table.

📘 In Simple Terms

It counts how many times each combination of two categories appears in your data.

```python
# Create a crosstab between frequency and platform
cross_tab = pd.crosstab(
    data["4.How often do you come across healthcare advertisements?"],
    data["5.Where do you mostly see or hear healthcare marketing?"]
)

# Plot stacked bar chart
cross_tab.plot(
    kind='bar',
    stacked=True,
    figsize=(8, 5),
    color=['lightcoral', 'lightblue', 'lightgreen', 'khaki', 'plum']
)
plt.title('Healthcare Advertisement Exposure by Platform')
plt.xlabel('How Often People See Healthcare Advertisements')
plt.ylabel('Number of Respondents')
plt.legend(title='Where People Mostly See or Hear Ads', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Display cross-tab table
print("\nCross-tab (Frequency vs Platform):")
print(cross_tab)
```

```python
cross_tab = pd.crosstab(data['3.Occupation'], data['2.Gender'])
cross_tab.plot(
    kind='bar',
    stacked=True,
    figsize=(7, 5),
    color=['lightcoral', 'lightblue', 'lightgreen']
)
plt.title('Gender Distribution by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Respondents')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

# Display cross-tab table
print("\nCross-tab (Gender vs Occupation):")
print(cross_tab)
```

```python
#  WHICH AD TYPE CATCHES ATTENTION MOST
data['6.Which type of healthcare advertisements catch your attention the most?'].value_counts().plot(
    kind='pie',
    color='mediumpurple',

    autopct="%1.1f%%"
)
plt.title('Types of Healthcare Ads That Catch Most Attention')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

#  TRUSTWORTHINESS SCORE
trust_scores = data['7.How trustworthy do you find healthcare advertisements?']

# Calculate key statistics
mean_trust = trust_scores.mean()
median_trust = trust_scores.median()
range_trust = trust_scores.max() - trust_scores.min()

print("Trustworthiness Score Summary:")
print(f"Mean: {mean_trust:.2f}")
print(f"Median: {median_trust}")
print(f"Range: {range_trust}")

# Histogram of trust scores
plt.figure(figsize=(6, 4))
plt.hist(trust_scores, bins=5, color='#7c59c9', edgecolor='black')
plt.title('Distribution of Trustworthiness Scores')
plt.xlabel('Trust Score (1–5)')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()
```

```python
# CROSS-ANALYSIS: TRUST BY AGE GROUP

plt.figure(figsize=(7, 4))
data.groupby('1.Age group')['7.How trustworthy do you find healthcare advertisements?'].mean().plot(
    kind='bar',
    color='lightgreen',
    edgecolor='black'
)
plt.title('Average Trust Score by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Trust Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# MOST EFFECTIVE MEDIUM (Q11)

plt.figure(figsize=(7, 4))
data['11.Which medium do you find most effective for healthcare marketing?'].value_counts().plot(
    kind='bar',
    color='cornflowerblue',
    edgecolor='black'
)
plt.title('Most Effective Medium for Healthcare Marketing')
plt.xlabel('Medium')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


#  CROSS-ANALYSIS: EFFECTIVE MEDIUM BY AGE GROUP

cross_tab_medium = pd.crosstab(
    data['1.Age group'],
    data['11.Which medium do you find most effective for healthcare marketing?']
)

cross_tab_medium.plot(
    kind='bar',
    stacked=True,
    figsize=(8, 5),
    color=['lightcoral', 'lightblue', 'lightgreen', 'khaki', 'plum']
)
plt.title('Most Effective Healthcare Marketing Medium by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.legend(title='Medium', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


#  DISPLAY CROSSTAB
print("\nCross-tab (Age Group vs Preferred Medium):")
print(cross_tab_medium)
```

The survey reveals that healthcare marketing today resonates most strongly with a younger, digitally-oriented audience, as social media platforms dominate both exposure and perceived effectiveness—particularly among respondents under 25, though television remains relevant for older adults. While the average trust in healthcare advertising is moderate, trust levels are notably higher among the youngest and oldest respondents, with those between 26 and 45 showing more skepticism. To maximize impact, campaigns should prioritize social and digital media, personalize messaging for different age groups, and focus on authenticity, clear value, and responsiveness. Aligning these practices with the latest industry trends—such as patient-centric and data-driven approaches, integrated cross-channel strategies, and influencer partnerships—ensures that healthcare marketing not only reaches audiences effectively but also builds lasting trust and engagement.
